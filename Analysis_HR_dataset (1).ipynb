```python
df = df[['WorkLifeBalance', 'Attrition']]
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
attrition_by_wlb = df.groupby("WorkLifeBalance")["Attrition"].mean() * 100
```

```python
attrition_table = attrition_by_wlb.reset_index()
attrition_table.columns = ["WorkLifeBalance", "Attrition Rate (%)"]
print(attrition_table)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(attrition_by_wlb.index, attrition_by_wlb.values, color='skyblue')
plt.xlabel("Work-Life Balance (1 = Low, 4 = High)")
plt.ylabel("Attrition Rate (%)")
plt.title("Impact of Work-Life Balance on Attrition Rate")
plt.xticks([1, 2, 3, 4])
plt.show()
```

```python
df = pd.read_csv('/content/HR_Analytics_Cleaned.csv')

df_travel = df[['BusinessTravel', 'Attrition']]
df_travel["Attrition"] = df_travel["Attrition"].map({"Yes": 1, "No": 0})
attrition_by_travel = df_travel.groupby("BusinessTravel")["Attrition"].mean() * 100
attrition_table = attrition_by_travel.reset_index()
attrition_table.columns = ["BusinessTravel", "Attrition Rate (%)"]
print(attrition_table)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(attrition_by_travel.index, attrition_by_travel.values, color='lightcoral')
plt.xlabel("Business Travel Frequency")
plt.ylabel("Attrition Rate (%)")
plt.title("Impact of Business Travel Frequency on Attrition Rate")
plt.xticks(rotation=15)
plt.show()
```

```python
# 3. Are employees with higher salaries more likely to stay?
```

```python
df_salary = df[['SalarySlab', 'Attrition']].copy()
df_salary["Attrition"] = df_salary["Attrition"].map({"Yes": 1, "No": 0})
attrition_by_salary = df_salary.groupby("SalarySlab")["Attrition"].mean() * 100
attrition_table = attrition_by_salary.reset_index()
attrition_table.columns = ["Salary Slab", "Attrition Rate (%)"]
print(attrition_table)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(attrition_by_salary.index, attrition_by_salary.values, color='purple')
plt.xlabel("Salary Slab")
plt.ylabel("Attrition Rate (%)")
plt.title("Impact of Salary Level on Attrition Rate")
plt.xticks(rotation=15)
plt.show()
```
